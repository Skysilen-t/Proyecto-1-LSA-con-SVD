```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import spacy
```

```python
df = pd.read_csv('bbc_data.csv')
df
```

```python
df.labels.astype('category')
```

```python
df = df[df['labels'] == 'politics'].reset_index()
df.sample(5)
print(f'Largo del dataset {len(df)}')
```

```python
import re

def limpiar_noticia(texto):
    #Eliminar etiquetas HTML
    texto = re.sub(r'<.*?>', ' ', texto)
    #Eliminar URLs y correos electrónicos
    texto = re.sub(r'http\S+|www\.\S+', ' ', texto)
    texto = re.sub(r'\S+@\S+', ' ', texto)
    #Reemplaza caracteres especiales con espacios
    texto = re.sub(r'[^\w\s]', ' ', texto)
    # 7. Eliminar espacios múltiples y convertir a minúsculas
    texto = re.sub(r'\s+', ' ', texto)
    return texto.lower().strip()
```

```python
etiquetas = df.index.astype(str).tolist()
#creamos una lista con todas las noticias aplicando el filtro de limpieza
documentos = [limpiar_noticia(text) for text in df['data']] 
```

```python
vectorizer_raw = CountVectorizer(
    lowercase=True,
    token_pattern=r"(?u)\b[^\W\d_]{3,}\b",
    min_df=1,
)

X_raw = vectorizer_raw.fit_transform(documentos)
vocabulario_raw = vectorizer_raw.get_feature_names_out()
A_raw = X_raw.toarray().T

print("Forma de X_raw (documentos x terminos):", X_raw.shape)
print("Forma de A_raw (terminos x documentos):", A_raw.shape)
print("Primeros terminos del vocabulario:", vocabulario_raw[:30])
```

```python
def tabla_resumen_distribucion(nombre, valores):
    return pd.DataFrame(
        [
            ["min", int(valores.min())],
            ["mediana", int(np.median(valores))],
            ["media", round(float(valores.mean()), 2)],
            ["max", int(valores.max())],
        ],
        columns=["estadistica", nombre],
    )


def resumen_vectorizacion(nombre, X, A, vocabulario):
    conteos_por_doc = np.asarray(X.sum(axis=1)).ravel()
    terminos_unicos_por_doc = np.asarray((X > 0).sum(axis=1)).ravel()
    total_celdas = X.shape[0] * X.shape[1]
    densidad = X.nnz / total_celdas

    tabla_corpus = pd.DataFrame(
        [
            ["numero_documentos", X.shape[0]],
            ["numero_terminos", X.shape[1]],
            ["forma_X_documentos_x_terminos", f"{X.shape[0]} x {X.shape[1]}"],
            ["forma_A_terminos_x_documentos", f"{A.shape[0]} x {A.shape[1]}"],
            ["entradas_no_cero", X.nnz],
            ["densidad", round(float(densidad), 4)],
            ["sparsity", round(float(1 - densidad), 4)],
        ],
        columns=["estadistica", nombre],
    )
    tabla_terminos_por_doc = tabla_resumen_distribucion(nombre, conteos_por_doc)
    tabla_terminos_unicos_por_doc = tabla_resumen_distribucion(nombre, terminos_unicos_por_doc)

    return (
        tabla_corpus,
        tabla_terminos_por_doc,
        tabla_terminos_unicos_por_doc,
        conteos_por_doc,
        terminos_unicos_por_doc,
    )

(
    tabla_corpus_raw,
    tabla_terminos_por_doc_raw,
    tabla_terminos_unicos_por_doc_raw,
    conteos_doc_raw,
    terminos_unicos_doc_raw,
) = resumen_vectorizacion(
    "corpus_sin_limpieza",
    X_raw,
    A_raw,
    vocabulario_raw,
)

display(tabla_corpus_raw)
display(tabla_terminos_por_doc_raw)
display(tabla_terminos_unicos_por_doc_raw)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(conteos_doc_raw, bins=10, color="tab:blue", alpha=0.85)
axes[0].set_title("Terminos totales por documento", fontsize=18)
axes[0].set_xlabel("Numero de terminos", fontsize=14)
axes[0].set_ylabel("Documentos", fontsize=14)
axes[0].grid(False)

axes[1].hist(terminos_unicos_doc_raw, bins=10, color="tab:green", alpha=0.85)
axes[1].set_title("Terminos unicos por documento", fontsize=18)
axes[1].set_xlabel("Numero de terminos unicos", fontsize=14)
axes[1].set_ylabel("Documentos", fontsize=14)
axes[1].grid(False)

plt.tight_layout()
plt.show()
```

```python
nlp_lista = spacy.load("en_core_web_sm", disable=["parser", "ner"])
nlp = spacy.load("en_core_web_sm")

stop_words_custom = [
    # Palabras funcionales en ingles
    "the", "a", "an", "and", "in", "to", "of", "for", "with", "on", "at", "by",
    "from", "up", "about", "into", "over", "after", "is", "are", "was", "were",
    "be", "been", "being", "have", "has", "had", "do", "does", "did", "will",
    "would", "shall", "should", "can", "could", "may", "might", "must", "it",
    "they", "them", "their", "this", "that", "these", "those", "which", "who",
    "whom", "whose", "what", "where", "when", "why", "how", "all", "any", "both",
    "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not",
    "only", "own", "same", "so", "than", "too", "very", "just", "now", "also", 
    "then", "well", "even", "because", "while", "since", "until", "thus", 
    "therefore", "however", "although", "already", "often", "always", "never", 
    "sometimes", "usually", "much", "many", "every", "first", "second", "next", 
    "previous", "given", "associated", "different", "exist", "existing", 
    "perform", "performed", "case", "word", "respect", "respective", 
    "respectively", "enough", "various", "various", "either", "neither", 'his'

    # Etiquetas frecuentes de las notas en ingles
    "def", "prop", "proof", "example", "observation", "question", "info",
    "note", "theorem", "lemma", "corollary", "remark", "definition"
]

stop_words_raw = set(ENGLISH_STOP_WORDS) | set(stop_words_custom)


def lematizar_lista(palabras):
    lemas = set()
    for palabra in palabras:
        doc = nlp_lista(palabra)
        for token in doc:
            lema = token.lemma_.lower().strip()
            if lema:
                lemas.add(lema)
    return lemas


# Filtramos dos veces: primero en la forma original y luego en su lema.
stop_words_lem = lematizar_lista(stop_words_raw)
stop_words_total = sorted(set(stop_words_raw) | stop_words_lem)


def lematizar_texto(texto):
    doc = nlp(texto)
    lemas = []
    for token in doc:
        texto_token = token.text.lower().strip()
        lema = token.lemma_.lower().strip()
        
        if not texto_token or not lema or token.is_space:
            continue
        # Mantenemos el filtro de longitud minima de 3 caracteres
        if len(texto_token) < 3 or len(lema) < 3:
            continue
        if texto_token in stop_words_total or lema in stop_words_total:
            continue
            
        lemas.append(lema)
    return " ".join(lemas)

documentos_lem = [lematizar_texto(texto) for texto in documentos]

vectorizer_final = TfidfVectorizer(
    lowercase=True,
    token_pattern=r"(?u)\b[^\W\d_]{3,}\b",
    stop_words=stop_words_total,
    min_df=4, #aparecer en más de 4 documentos
    max_df=0.8, #aparecer en menos del 80%
    ngram_range=(1, 2), #busqueda de unigramas y bigramas
)

'''
Creamos la matriz TF-IDF
matriz donde cada fila es un documento y cada columna es una palabra
asignando puntaje de importancia
'''
X_final = vectorizer_final.fit_transform(documentos_lem)

#diccionario de filtros
vocabulario_final = vectorizer_final.get_feature_names_out()
#matriz dispersa
A_final = X_final.toarray().T #Buscamos la transpuesta para obtener la matriz término-documento

#vectorizador de las palabras que usamos para la matriz TF-IDF
vectorizer_final_counts = CountVectorizer(vocabulary=vocabulario_final)
#Conteo de palabras
X_final_counts = vectorizer_final_counts.transform(documentos_lem)
#array donde cada número representa la frecuencia total de la palabra
frecuencias_final = np.asarray(X_final_counts.sum(axis=0)).ravel()

print(f'Largo del vocabulario final {len(vocabulario_final)}')
```

## P3:

Primero creamos nuestra lista de stopwords que es una unión entre las ENGLISH_STOP_WORDS de sklearn.feature_extraction.text y stop_words_custom(una lista de algunas palabras que decidimos agregar). A esta lista lo lematizamos con el modelo "en_core_web_sm" de nlp de la librería spacy. Creamos una lista (documentos_lem) con todos los documentos de las noticias, esta lista le aplicamos igual la lematización de spacy. Luego, creamos nuestra matriz TF-IDF, la cual asigna un puntaje de importancia a cada palabra, el filtro para la creación es de palabras que no esten en los stopwords seleccionados, que aparecen en más de 4 documentos y al menos en el 80% de ellos (escogimos este filtro ya que estamos trabajando con más de 400 documentos), de unigramas y bigramas, estos filtros estrictos nos permiten obtener las palabras que más aportan información y descartar las que no. Luego, Obtenemos nuestro vocabulario de 4558 y creamos nuestra matriz término-documento (A_final). Luego creamos un vectorizador para poder obtener el conteo de las palabras y obtenemos nuestro array con la frecuencia de las palabras.

```python
print("Forma de A_final (terminos x documentos):", A_final.shape)
```

## P4

Utilizamos scikit-learn que entrega una matriz con documentes en las filas y términos en las columnas por lo tanto obtenemos su transpuesta para poder trabajo con la matriz con términos en las filas y documentes en las columnas. La Forma de la matriz (terminos x documentos): (4558, 417). Una entrada $A_{ij}$ representa el puntaje TF-IDF (Frecuencia de Término - Frecuencia Inversa de Documento) del término $i$ en el documento $j$. Este valor numérico indica qué tan importante o relevante es el término $i$ específicamente para el documento $j$

```python
indices_frecuentes_final = np.argsort(frecuencias_final)[::-1][:30]
terminos_frecuentes_final = vocabulario_final[indices_frecuentes_final]
conteos_frecuentes_final = frecuencias_final[indices_frecuentes_final]

fig, ax = plt.subplots(figsize=(11, 5.2))
ax.bar(terminos_frecuentes_final, conteos_frecuentes_final, alpha=0.85)
ax.set_title("Terminos mas frecuentes despues de limpiar", fontsize=18)
ax.set_xlabel("Termino", fontsize=14)
ax.set_ylabel("Frecuencia en documentos", fontsize=14)
ax.grid(False)
ax.tick_params(axis="x", rotation=70, labelsize=14)
plt.tight_layout()
plt.show()
```

## P5:
Los valores singulares representan la cantidad de información(varianza) que captura cada tema, el primer valor es el más dominante, en cambio los últimos son temas raros. Los vectores singulares izquierdos representan la relación de las palabras con los temas y los vectores derechos representan la relación de los temas con los documentos.

```python
U_raw, s_raw, VT_raw = np.linalg.svd(A_final, full_matrices=False)

print("Primeros valores singulares")
print(np.round(s_raw[:10], 3))

plt.figure(figsize=(10, 5))
plt.plot(s_raw, marker="o")
plt.title("Valores singulares:", fontsize=16)
plt.xlabel("Indice", fontsize=14)
plt.xticks(np.arange(0, len(s_raw) + 1, step=20), rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylabel("Valor singular", fontsize=14)
plt.show()
```

## P6:
Al graficar los valores singulares, se observa una caída pronunciada al inicio. Elegimos $k = 80$ basándonos en el método del codo, ya que a partir de este componente la magnitud de los valores singulares se estabiliza, indicando que el resto de las componentes representan mayoritariamente ruido en lugar de estructura semántica relevante del corpus.

```python
indices = [0, 1]

D_2 = np.diag(s_raw[indices]) @ VT_raw[indices, :]
T_2 = U_raw[:, indices] @ np.diag(s_raw[indices])
```

## P7:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.scatter(D_2[0, :], D_2[1, :], alpha=0.6)
ax1.set_title('Representación 2D de Documentos\n(Componentes 1 y 2)')
ax1.set_xlabel('Componente 1')
ax1.set_ylabel('Componente 2')
ax1.grid(True, linestyle='--', alpha=0.5)

ax2.scatter(T_2[:, 0], T_2[:, 1], alpha=0.4, color='crimson', marker='.')
ax2.set_title('Representación 2D de Términos\n(Componentes 1 y 2)')
ax2.set_xlabel('Componente 1')
ax2.set_ylabel('Componente 2')
ax2.grid(True, linestyle='--', alpha=0.5)

distancias = np.linalg.norm(T_2, axis=1)
indices_lejanos = np.argsort(distancias)[::-1][:10]

for i in indices_lejanos:
    ax2.annotate(vocabulario_final[i], (T_2[i, 0], T_2[i, 1]), fontsize=9)

plt.tight_layout()
plt.show()
```

Con respecto a los Documentos vemos varios puntos acumulados, es decir, tratan sobre temas similares, pero al mismo tiempo vemos que están bien dispersos, es decir, que hablan sobre temas distintos. En la gráfica de términos vemos una acumulación de palabras hacia los extremos derechos, entonces, la mayoría de los términos suelen aparecer juntas o en contextos similares a lo largo de los documentos, y solo algunas pocas se acercan al centro, que son las palabras más comunes en los documentos.

```python
# Convertimos tu vocabulario final a lista para buscar fácilmente
vocab_lista = list(vocabulario_final)

print("="*50)
print(" RESULTADOS PARA LA PREGUNTA 8 (EXPLORACIÓN SEMÁNTICA)")
print("="*50, "\n")

# --- EJEMPLO 1: Significado Temático del Componente 1 ---
# Buscamos qué palabras están en los extremos del Eje X
indices_c1 = np.argsort(T_2[:, 0])
palabras_izq = [vocab_lista[i] for i in indices_c1[:6]]
palabras_der = [vocab_lista[i] for i in indices_c1[-6:]]

print("1. COMPONENTE CON SIGNIFICADO TEMÁTICO (COMPONENTE 1)")
print(f"   -> Extremo Negativo agrupa: {palabras_izq}")
print(f"   -> Extremo Positivo agrupa: {palabras_der}\n")


# --- EJEMPLO 2: Palabras Cercanas (Relación Semántica) ---
# Calculamos la similitud del coseno entre todas las palabras
similitud_palabras = cosine_similarity(T_2)

# Usamos 'blair' ya que sabemos que es un texto político británico
palabra_test = "blair" 

if palabra_test in vocab_lista:
    idx_test = vocab_lista.index(palabra_test)
    # Buscamos las 5 palabras más similares (excluyendo la palabra misma)
    idx_cercanos = np.argsort(similitud_palabras[idx_test])[::-1][1:6]
    palabras_cercanas = [vocab_lista[i] for i in idx_cercanos]
    
    print(f"2. PALABRAS CERCANAS A '{palabra_test.upper()}'")
    print(f"   -> El modelo la asocia fuertemente con: {palabras_cercanas}\n")
else:
    print(f"2. La palabra '{palabra_test}' no está en el vocabulario.\n")


# --- EJEMPLO 3: Documentos Cercanos (Agrupamiento Temático) ---
# Transponemos D_2 para comparar los documentos usando similitud del coseno
similitud_docs = cosine_similarity(D_2.T)

# Buscaremos qué documentos se parecen al Documento 10 (puedes cambiar este número)
doc_test_idx = 10 
idx_docs_cercanos = np.argsort(similitud_docs[doc_test_idx])[::-1][1:4]

print(f"3. DOCUMENTOS CERCANOS AL DOCUMENTO {doc_test_idx}")
print(f"   -> Los más similares son los índices: {idx_docs_cercanos}")

# Imprimimos los primeros 120 caracteres para ver de qué tratan
print(f"\n   [Extracto Doc {doc_test_idx} (Original)]: {documentos[doc_test_idx][:120]}...")
for i in idx_docs_cercanos:
    print(f"   [Extracto Doc {i} (Similar)] : {documentos[i][:120]}...")
```
